```python
# Available operators: Greater(>), Less(<),Equal(==)    --   one = declares, value ==compares
# Greater or Equal(>=), Less or Equal (<=)

if 5 > 4:
    print('5 is greater than 4')
    
if 5 < 4:
    print('5 is less than 4') #Skip action if boolean returns false

if 5 == 5:
    print('5 is equal to 5')
    
if 5 + 5 > 9:
    print('10 is greater than 9')
    
if 10 != 9:
    print('10 is not equal to 9')
    
# Truth Tree:
# T and F = F
# T and T = T
# T or F = T
# F or T = T
# F or F = F
```

##### 'is' keyword

```python
list1 = [1,2,3,4]
list2 = [1,2,3,4]
#True
list1 == list2
#False
list1 is list2

list1_copy = list1
print(list1_copy)

print(list1_copy is list1)
list1.append(5)

print(list1_copy)
print(list1)
print(list2)
```

##### 'in' keyword

```python
#membership check
character = 'Neo'
print('N' in character)
#case sensative
print('n' in character)

print('neo' in 'neo')
#out of order
print('one' in 'neo')

if 'N' in 'Neo':
    print('N in Neo')
```

```python
for character in ['o', 'n', 'e', 'a', 'b']:        #List comprehension, covering later in week
    if character in 'neo':
        print(f'{character} true')
```

##### 'not in' keyword'

```python
if 'z' not in character:
    print('no z in Neo')
    
if 'n' not in character:
    print('no n in Neo')
    
if 'N' not in character:
    print('There is an N in Neo')
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didn’t
# An else after a for runs if the for didn’t break

for num in range (1,101):
    count = 0
    for i in range(2, (num // 2 + 1)):  #Start at 2 b/c 1 is not a prime number
        if num % i == 0:
            count = count + 1
            break
    if count == 0 and num != 1:
        print(' %d' %num)
```

# Exercise 3 <br>
<p>Take in a users input for their age, if they are younger than 18 print kids, if they're 18 to 65 print adults, else print seniors</p>

```python
while True:
    try:
        age = int(input('What is your age?\n')) #Prevents non int input breaking code
        break
    except:
        print('Error, enter age in digits...')
        
if age < 18:
    print('kids')
elif age >= 18 and age <= 65:
    print('adults')
elif age > 65:
    print('seniors')
```
